```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
os.chdir("C:/Users/Admin/Documents/Porfolio/SQL Tableau")
df = pd.read_csv('CarRentalDataV1.csv')
df = pd.DataFrame(df)
n = len(df)
n # 5851 cars
df.columns
```

```python
#Exploration location
#print(np.unique(df["location.country"])) # only US car
#print(np.unique(df["location.state"])) # 43/50 states
len(np.unique(df["location.city"])) #971 cities
```

```python
#Exploration vehicle type
#df["fuelType"] = df["fuelType"].astype("str")
#np.unique(df["fuelType"]) # 'DIESEL', 'ELECTRIC', 'GASOLINE', 'HYBRID', 'not_known'
df_type = df.groupby(['fuelType'])['fuelType'].count().sort_values(ascending=False).reset_index(name='count')

df_type["percent"] = (df_type["count"]/df_type["count"].sum())*100
df_type.replace('nan', 'UNKNOWN')
```

```python
#Exploration vehical brand
df_brand = df.groupby(['vehicle.make'])['vehicle.make'].count().sort_values(ascending=False).reset_index(name='count')
df_brand["percent"] = (df_brand["count"]/df_brand["count"].sum())*100
df_brand.replace('nan', 'UNKNOWN')
```

```python
#Exploration vehical style
df_style = df.groupby(['vehicle.type'])['vehicle.type'].count().sort_values(ascending=False).reset_index(name='count')
df_style["percent"] = (df_style["count"]/df_style["count"].sum())*100
df_style.replace('nan', 'UNKNOWN')
```

```python
#Exploration state
df_state = df.groupby(['location.state'])['location.state'].count().sort_values(ascending=False).reset_index(name='count')
df_state["percent"] = (df_state["count"]/df_state["count"].sum())*100
df_state.replace('nan', 'UNKNOWN')
```

```python
#Exploration city
df_city = df.groupby(['location.city'])['location.city'].count().sort_values(ascending=False).reset_index(name='count')
df_city["percent"] = (df_city["count"]/df_city["count"].sum())*100
df_city.replace('nan', 'UNKNOWN')
```

```python
#Exploration year
df_year = df.groupby(['vehicle.year'])['vehicle.year'].count().sort_values(ascending=False).reset_index(name='count')
df_year["percent"] = (df_year["count"]/df_year["count"].sum())*100
df_year.replace('nan', 'UNKNOWN')
```

```python
### add_index
list = []
for i in range(n):
    index = "C" + str(i).zfill(4)
    list.append(index)
```

```python
### add and rename columns
df["car_index"] = list

df = df.rename(columns={'location.city': 'location_city',
                   'location.country': 'location_country',
                   'location.latitude': 'location_latitude',
                   'location.longitude': 'location_longitude',
                   'owner.id': 'owner_id',
                   'rate.daily': 'rate_daily',
                   'vehicle.make': 'vehicle_make',
                   'vehicle.model': 'vehicle_model',
                   'vehicle.type': 'vehicle_type',
                   'vehicle.year': 'vehicle_year',
                   'location.state': 'location_state'
                   })

df.columns
```

```python
############ split data
car_location = df[["car_index","location_country","location_state","location_city",
                   "location_latitude","location_longitude"]]

car_admin = df[["car_index","fuelType","owner_id",
                "vehicle_make","vehicle_model","vehicle_type", "vehicle_year"]]

car_performance = df[["car_index","rating","renterTripsTaken","reviewCount","rate_daily"]]
```

```python
car_location.to_csv("C:/Users/Admin/Documents/Porfolio/SQL Tableau/location.csv")
car_admin.to_csv("C:/Users/Admin/Documents/Porfolio/SQL Tableau/admin.csv")  
car_performance.to_csv("C:/Users/Admin/Documents/Porfolio/SQL Tableau/performance.csv")    
```
